```python
a = os.listdir('C:/Users/Acer/Downloads/прога')
from Bio import SeqIO
from itertools import chain
os.chdir('C:/Users/Acer/Downloads/прога')
def main():
    c=0
    for i in a:  
        h = open(i, "rb")
        c+=1
        for record in SeqIO.parse(h, "abi"):
            
            pp = list(record.seq)

            input_h = pp
            kakoyto_dictionary = {
    'A': 'Ala',
    'C': 'Cys',
    'D': 'Asp',
    'E': 'Glu',
    'F': 'Phe',
    'G': 'Gly',
    'H': 'His',
    'I': 'Ile',
    'K': 'Lys',
    'L': 'Leu',
    'M': 'Met',
    'N': 'Asn',
    'P': 'Pro',
    'Q': 'Gln',
    'R': 'Arg',
    'S': 'Ser',
    'T': 'Thr',
    'V': 'Val',
    'W': 'Trp',
    'Y': 'Tyr'
            }
            output = ''
            for element in input_h:
                output += kakoyto_dictionary[element]
            #print(output)
            r = str(output)
        print(r)
            
main()
```
